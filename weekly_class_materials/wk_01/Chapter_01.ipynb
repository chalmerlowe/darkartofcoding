<h1>Operators</h1>

| Operator   | Operation     | Example   | Result   |
|:----------:|:--------------|----------:|---------:|
|**          |Exponent       | 2 ** 4    | 16       |
|*           |Multiplication | 3 * 4     | 12       |
|/           |Division       | 12 / 4    | 3        |
|+           |Addition       | 3 + 3     | 6        |
|-           |Subtraction    | 8 - 4     | 4        |


The order of operations matches the order used in mathematics.
Parentheticals can be used to override precedence.


```python
>>> 1 + 2 / 3
```

```python
>>> (1 + 2) / 3
```

```python
>>> ((4 - 1) * (4 + 1)) / (7 - 2)  
```
