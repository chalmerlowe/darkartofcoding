<h1>'AND' Truth Table</h1>

Operator 1 |     | Operator 2 | Results in
-----------|-----|------------|-------
True       | and | True       | True
True       | and | False      | False
False      | and | True       | False
False      | and | False      | False


```python
>>> True and True
```

```python
>>> False and True
```

<h1>'OR' Truth Table</h1>

Operator 1 |     | Operator 2 | Results in
-----------|-----|------------|-------
True       | or  | True       | True
True       | or  | False      | True
False      | or  | True       | True
False      | or  | False      | False

```python
>>> False or False
```

```python
>>> True or False
```

```python
>>> not False
```

```python
>>> not 13 + 42 == 0
#   not (13 + 42 == 0)
```

```python
# mixing things up
>>> (13 < 42) and (3 > 1)
```
